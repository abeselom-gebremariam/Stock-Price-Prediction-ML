```python
import sys
sys.path.append('../src')  # So Python can find your src/ folder

from data_processing import download_stock_data, split_data, scale_data, prepare_sequences, prepare_test_data
from model import build_lstm_model

# Step 1: Load data
data = download_stock_data()
print(data) 
# Step 2: Visualize
import matplotlib.pyplot as plt

ma_100_days = data.Close.rolling(100).mean()
ma_200_days = data.Close.rolling(200).mean()

ma_100_days = data.Close.rolling(100).mean()
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r', label='MA 100')
plt.plot(data.Close, 'g', label='Closing Price')
plt.legend()
plt.show()

plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r', label='MA 100')
plt.plot(ma_200_days, 'b', label='MA 200')
plt.plot(data.Close, 'g', label='Closing Price')
plt.legend()
plt.show()

# Step 3: Preprocess
data_train, data_test = split_data(data)
print(f"Data train size: {data_train.shape[0]}")
print(f"Data test size: {data_test.shape[0]}")

data_train_scaled, scaler = scale_data(data_train)

x, y = prepare_sequences(data_train_scaled)

# Step 4: Build and train model
model = build_lstm_model((x.shape[1], 1))
model.fit(x, y, epochs=50, batch_size=32, verbose=1)

# Summary
model.summary()

# Step 5: Prepare the test data
x_test, y_test, data_test_scaled = prepare_test_data(data_train, data_test, scaler)

# Step 6: Make predictions
y_predict = model.predict(x_test)

# Step 7: Inverse scaling
scale = 1 / scaler.scale_[0]
y_predict = y_predict * scale
y_test = y_test * scale

# Step 8: Visualize the results
plt.figure(figsize=(10, 8))
plt.plot(y_predict, 'r', label='Predicted Price')
plt.plot(y_test, 'g', label='Original Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

model.save('../models/stock_predictions_model.keras')
```
